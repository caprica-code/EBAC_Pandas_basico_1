## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui


notas = pd.DataFrame(np.random.randint(0, 10, size=(100, 1)), columns=['% acerto MAT'])


notas['% acerto MAT'] = (df_mat['Qt_acertos'] / 24) * 100
notas['% acerto POR'] = (df_por['Qt_acertos'] / 18) * 100
notas['% acerto GEO'] = (df_geo['Qt_acertos'] / 8) * 100
notas['% acerto ING'] = (df_ing['Qt_acertos'] / 8) * 100
notas['% acerto HIS'] = (df_his['Qt_acertos'] / 8) * 100
notas['% acerto FIS'] = (df_fis['Qt_acertos'] / 12) * 100
notas['% acerto QUI'] = (df_qui['Qt_acertos'] / 12) * 100
notas = notas.round(2)
notas
````

````python
# 2) Seu código aqui
df_soma_nota = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
df_soma_nota.columns = ['Total de Acertos']
df_soma_nota
````

````python
# 3) Seu código aqui
df_percent_aluno = (df_soma_nota['Total de Acertos'] / 100)
df_percent_aluno = pd.DataFrame(df_percent_aluno)
df_percent_aluno.columns = ['% de acertos Totais'] 
df_percent_aluno
````

````python
# 4) Seu código aqui
df_aprovados = df_soma_nota['Total de Acertos'] > 44
df_aprovados = pd.DataFrame(df_aprovados)
df_aprovados.columns = ['Aprovados']
quantidade_aprovados = df_aprovados['Aprovados'].sum()
print("\nQuantidade de alunos aprovados:", quantidade_aprovados, '\n')
print("DataFrame de aprovados:")
df_aprovados



````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
#código dos desqualificados em MAT, FIS e QUI
print("\nData Frame de alunos reprovados em MAT, FIS e QUI:\n")
df_desq = (df_mat['Qt_acertos'] < 1) + (df_fis['Qt_acertos'] < 1) + (df_qui['Qt_acertos'] < 1)
df_desq = pd.DataFrame(df_desq)
df_desq.columns = ['Reprovados']
df_desq
````

````python
#código mostrando a quantidade de alunos desqualificados em MAT, FIS e QUI
quantidade_reprovados = df_desq['Reprovados'].sum()
print('=-'*17)
print("Quantidade de alunos aprovados:", quantidade_reprovados)
print('=-'*17)
````

````python
#código mostrando a média dos alunos em HIS e GEO
print("\nData Frame com a média dos alunos em HIS e GEO:\n")
df_combined = pd.concat([df_geo, df_his], axis=1)
df_combined.columns = ['Qt_acertos_geo', 'Qt_acertos_his']
df_combined['Media'] = df_combined.mean(axis=1)
df_combined
````

````python
#código contendo o dataframe dos alunos que que foram aprovados seguindo duas métricas
# média em todas materias maior que 45 e não zerar em MAT,FIS e QUI

#utilizando a NEGAÇÃO '~' para inverter e colocando todos criterios dentro de uma msm equação utilizando o criterio OU '|'
df_desq_invert = ~((df_mat['Qt_acertos'] < 1) | (df_fis['Qt_acertos'] < 1) | (df_qui['Qt_acertos'] < 1))
df_desq_invert = pd.DataFrame(df_desq_invert)
df_desq_invert.columns = ['Aprovados']

#utilizando o criterio E '&' para somar dois df que contenham apenas o msm resultado true
df_aprove_geral = df_aprovados['Aprovados'] & df_desq_invert['Aprovados']
df_aprove_geral = pd.DataFrame(df_aprove_geral)
df_aprove_geral.columns = ['Geral Aprovados']

qtd_aprov_geral = df_aprove_geral['Geral Aprovados'].sum()
print('=-'*17)
print("Quantidade de alunos aprovados:", qtd_aprov_geral)
print('=-'*17)

df_aprove_geral
````
